```python
def clean_grocery_area(value):
    """
    Clean grocery area values by converting valid numbers to float
    and handling various edge cases
    """
    if pd.isna(value) or value in ['', ' ', 'nil', 'Not Specified', 'To Be Determined', 
                                  'TBD', 'missing', 'unknown', '?', '###', '...', '--',
                                  'Not Specified', 'To Be Determined']:
        return np.nan
    
    # Remove quotes if present
    if isinstance(value, str) and value.startswith('"') and value.endswith('"'):
        value = value[1:-1]
    
    # Handle negative values and special codes
    if value in ['-1', '-999']:
        return np.nan
    
    # Try to convert to float
    try:
        return float(value)
    except (ValueError, TypeError):
        return np.nan

# Apply the cleaning function
train1['groceryArea'] = train1['groceryArea'].apply(clean_grocery_area)
```

```python
def clean_frozen_area(value):
    """
    Clean frozen area values by converting valid numbers to float
    and handling various edge cases
    """
    if pd.isna(value) or value in ['', ' ', '?', '###', 'missing', 'Not Available', 
                                  'To Be Determined', 'TBD', '-1', '...', '--', 'nil']:
        return np.nan
    
    # Convert to string for processing
    value_str = str(value).strip()
    
    # Handle specific problematic values
    if value_str.lower() in ['?', '###', 'missing', 'not available', 'to be determined', 
                           'tbd', '-1', '...', '--', 'nil']:
        return np.nan
    
    # Try to convert to float
    try:
        result = float(value_str)
        # Optional: Add validation for reasonable range
        if result < 0 :  
            return np.nan
        return result
    except (ValueError, TypeError):
        return np.nan

# Apply the cleaning function
train1['frozenArea'] = train1['frozenArea'].apply(clean_frozen_area)
```

```python
def clean_meat_area(value):
    """
    Clean meat area values by converting valid numbers to float
    and handling various edge cases
    """
    if pd.isna(value) or value in ['', ' ', '?', '###', 'missing', 'Not Available', 
                                  'To Be Determined', 'TBD', '-1', '...', '--', 'nil',
                                  'Not Specified', 'Data Missing', 'unknown']:
        return np.nan
    
    # Convert to string for processing
    value_str = str(value).strip()
    
    # Remove quotes if present
    if value_str.startswith('"') and value_str.endswith('"'):
        value_str = value_str[1:-1]
    
    # Handle specific problematic values
    if value_str.lower() in ['?', '###', 'missing', 'not available', 'to be determined', 
                           'tbd', '-1', '...', '--', 'nil', 'not specified', 'data missing',
                           'unknown']:
        return np.nan
    
    # Try to convert to float
    try:
        result = float(value_str)
        # Optional: Add validation for reasonable range
        if result < 0 :
            return np.nan
        return round(result, 2)  # Round to 2 decimal places for consistency
    except (ValueError, TypeError):
        return np.nan

# Apply the cleaning function
train1['meatArea'] = train1['meatArea'].apply(clean_meat_area)
```

```python
from sklearn.impute import KNNImputer

# Method 4: K-Nearest Neighbors imputation
def impute_with_knn(data, n_neighbors=5):
    """Use KNN to impute missing values"""
    imputer = KNNImputer(n_neighbors=n_neighbors)
    imputed_array = imputer.fit_transform(data)
    
    imputed_df = pd.DataFrame(imputed_array, 
                             columns=data.columns, 
                             index=data.index)
    return imputed_df

# Apply KNN imputation
train1[['groceryArea', 'frozenArea', 'meatArea']] = impute_with_knn(train1[['groceryArea', 'frozenArea', 'meatArea']])
```

```python
train1['grossWeight'].value_counts()
```

```python
def clean_weight(value):
    if pd.isna(value):
        return np.nan
    
    s = str(value).strip().lower()
    
    # Missing placeholders
    if s in ["not available", "nil", "?", "unknown", "to be determined", "tbd", "--"]:
        return np.nan
    
    # Remove "data:", "val:", "amount:", "measure:"
    for prefix in ["data:", "val:", "amount:", "measure:"]:
        if s.startswith(prefix):
            s = s.replace(prefix, "").strip()
    
    # Handle units
    if "kg" in s:
        return float(s.replace("kg", "").strip())
    elif "grams" in s:
        return float(s.replace("grams", "").strip()) / 1000
    elif "lbs" in s:
        return float(s.replace("lbs", "").strip()) * 0.453592
    elif "oz" in s:
        return float(s.replace("oz", "").strip()) * 0.0283495
    
    # Try convert to float
    try:
        val = float(s)
        return val if val > 0 else np.nan 
    except:
        return np.nan

train1["grossWeight"] = train1["grossWeight"].apply(clean_weight)
```

## `netWeight` column

```python
train1['netWeight'].value_counts()
```

```python
import pandas as pd
import numpy as np
import re

def clean_net_weight(value):
    if pd.isna(value):
        return np.nan
    
    s = str(value).strip().lower()
    

    if s in ["not available", "nil", "?", "unknown", "to be determined", "tbd", "--", "###"]:
        return np.nan
    

    for prefix in ["data:", "val:", "amount:", "measure:"]:
        if s.startswith(prefix):
            s = s.replace(prefix, "").strip()
    

    match = re.findall(r"[-+]?\d*\.?\d+", s)
    if not match:
        return np.nan
    
    try:
        val = float(match[0])
    except:
        return np.nan
    
    
    if val <= 0:
        return np.nan
    

    if "kg" in s:
        return val
    elif "grams" in s or "g " in s or s.endswith("g"):
        return val / 1000
    elif "lbs" in s or "lb" in s:
        return val / 2.205
    elif "oz" in s:
        return val / 35.274
    else:
        return val  

train1["netWeight"] = train1["netWeight"].apply(clean_net_weight)
```

```python
def simple_weights_cleaner(df, column_name='weights'):
    """
    Simpler approach focusing on pattern matching
    """
    df = df.copy()
    
    # Extract numeric patterns
    def extract_numeric_patterns(text):
        if pd.isna(text):
            return np.nan, np.nan
        
        text = str(text)
        
        # Find all numeric values (including decimals and negatives)
        numbers = re.findall(r'-?\d+\.?\d*', text)
        numbers = [float(num) for num in numbers if abs(float(num)) < 1000 and float(num) != -999.0]
        
        if len(numbers) >= 2:
            return numbers[0], numbers[1]  # Assume first two are gross and net
        elif len(numbers) == 1:
            return numbers[0], np.nan
        else:
            return np.nan, np.nan
    
    extracted = df[column_name].apply(extract_numeric_patterns)
    df['gross_weight'] = extracted.apply(lambda x: x[0])
    df['net_weight'] = extracted.apply(lambda x: x[1])
    
    return df.drop(column_name, axis=1)


train2 = simple_weights_cleaner(train2, 'weights')
```

```python
def clean_gross_weight(df, column_name='grossWeight'):
    """
    Clean the grossWeight column by handling various data quality issues
    """
    # Create a copy to avoid modifying original data
    df_clean = df.copy()
    
    # Convert to string first to handle mixed types
    df_clean[column_name] = df_clean[column_name].astype(str)
    
    # Remove common prefixes and extract numeric values
    prefixes = ['amount:', 'val:', 'data:', 'measure:']
    for prefix in prefixes:
        df_clean[column_name] = df_clean[column_name].str.replace(prefix, '', regex=False)
    
    # Remove parentheses and other special characters
    df_clean[column_name] = df_clean[column_name].str.replace(r'[()]', '', regex=True)
    
    # Handle unit conversions - extract numeric part and convert to consistent units (kg)
    def convert_units(value):
        try:
            # Check for oz (ounces to kg)
            if 'oz' in value.lower():
                num = float(''.join(filter(lambda x: x.isdigit() or x == '.', value)))
                return num * 0.0283495  # Convert oz to kg
            
            # Check for lbs (pounds to kg)
            elif 'lbs' in value.lower():
                num = float(''.join(filter(lambda x: x.isdigit() or x == '.', value)))
                return num * 0.453592  # Convert lbs to kg
            
            # Check for grams to kg
            elif 'grams' in value.lower() or 'gram' in value.lower():
                num = float(''.join(filter(lambda x: x.isdigit() or x == '.', value)))
                return num / 1000  # Convert grams to kg
            
            # Check for kg (already in correct unit)
            elif 'kg' in value.lower():
                return float(''.join(filter(lambda x: x.isdigit() or x == '.', value)))
            
            # Regular numeric value
            else:
                return float(value)
                
        except (ValueError, TypeError):
            return np.nan
    
    # Apply unit conversion
    df_clean[column_name] = df_clean[column_name].apply(convert_units)
    
    # Handle negative values and outliers
    df_clean[column_name] = df_clean[column_name].apply(
        lambda x: np.nan if (pd.isna(x) or x <= 0 ) else x
    )
    
    # Handle specific problematic values
    df_clean[column_name] = df_clean[column_name].replace({
        -999.0: np.nan,
        -1.0: np.nan,
        -0.1: np.nan
    })
    
    return df_clean

train4 = clean_gross_weight(train4, 'grossWeight')
```

```python
def clean_net_weight(df, column_name='netWeight'):
    """
    Clean the netWeight column by handling various data quality issues
    """
    # Create a copy to avoid modifying original data
    df_clean = df.copy()
    
    # Convert to string first to handle mixed types
    df_clean[column_name] = df_clean[column_name].astype(str)
    
    # Remove common prefixes and extract numeric values
    prefixes = ['amount:', 'val:', 'data:', 'measure:']
    for prefix in prefixes:
        df_clean[column_name] = df_clean[column_name].str.replace(prefix, '', regex=False)
    
    # Remove parentheses, brackets, and other special characters
    df_clean[column_name] = df_clean[column_name].str.replace(r'[()\[\]]', '', regex=True)
    
    # Handle unit conversions - extract numeric part and convert to consistent units (kg)
    def convert_units(value):
        try:
            # Check for oz (ounces to kg)
            if 'oz' in value.lower():
                num = float(''.join(filter(lambda x: x.isdigit() or x == '.' or x == '-', value)))
                return num * 0.0283495  # Convert oz to kg
            
            # Check for lbs (pounds to kg)
            elif 'lbs' in value.lower() or 'lb' in value.lower():
                num = float(''.join(filter(lambda x: x.isdigit() or x == '.' or x == '-', value)))
                return num * 0.453592  # Convert lbs to kg
            
            # Check for grams to kg
            elif 'grams' in value.lower() or 'gram' in value.lower():
                num = float(''.join(filter(lambda x: x.isdigit() or x == '.' or x == '-', value)))
                return num / 1000  # Convert grams to kg
            
            # Check for kg (already in correct unit)
            elif 'kg' in value.lower():
                return float(''.join(filter(lambda x: x.isdigit() or x == '.' or x == '-', value)))
            
            # Regular numeric value
            else:
                return float(value)
                
        except (ValueError, TypeError):
            return np.nan
    
    # Apply unit conversion
    df_clean[column_name] = df_clean[column_name].apply(convert_units)
    
    # Handle negative values and outliers
    df_clean[column_name] = df_clean[column_name].apply(
        lambda x: np.nan if (pd.isna(x) or x <= 0 or x > 1000) else x
    )
    
    # Handle specific problematic values
    df_clean[column_name] = df_clean[column_name].replace({
        -999.0: np.nan,
        -1.0: np.nan,
        -0.1: np.nan
    })
    
    # Handle common missing value representations
    missing_indicators = ['nan', 'missing', 'not specified', 'tbd', 'data missing', 
                         'not available', 'nil', '--', '###', '?', 'unknown']
    df_clean[column_name] = df_clean[column_name].apply(
        lambda x: np.nan if str(x).lower() in missing_indicators else x
    )
    
    return df_clean


train4 = clean_net_weight(train4, 'netWeight')
```
